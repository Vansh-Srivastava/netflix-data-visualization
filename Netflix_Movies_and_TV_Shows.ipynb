```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Upload CSV
uploaded = files.upload()
df = pd.read_csv("netflix_titles.csv")

# Preview data
df.head()
```

```python
# Clean and convert 'date_added' column
df['date_added'] = pd.to_datetime(df['date_added'].astype(str).str.strip(), errors='coerce')

# Extract year
df['year_added'] = df['date_added'].dt.year
```

```python
# Plot number of titles added per year
df['year_added'].value_counts().sort_index().plot(kind='line', marker='o', color='green')
plt.title("Netflix Titles Added Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.grid(True)
plt.show()
```

```python
# Movies vs TV Shows
df['type'].value_counts().plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title("Movies vs TV Shows on Netflix")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

```python
# Handle missing countries
df['country'].fillna('Unknown', inplace=True)

# Plot top countries
df['country'].value_counts().head(10).plot(kind='bar', color='purple')
plt.title("Top 10 Countries by Netflix Content")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```
